# Covariance & Correlation

**Dataset:** 15 students — hours studied, exam score, and hours of sleep.

Expected relationships:
- Hours studied vs Exam score → positive correlation
- Hours studied vs Hours slept → negative correlation (students who study more sleep less)

**Formulas:**
- Covariance: `cov(X,Y) = Σ(xi - x̄)(yi - ȳ) / (n-1)`
- Pearson correlation: `r = cov(X,Y) / (std_X * std_Y)`  (always between -1 and 1)

```python
import numpy as np
import matplotlib.pyplot as plt

# 15 students: hours studied, exam score, hours slept
hours_studied = np.array([2, 3, 1, 5, 4, 6, 3, 7, 8, 5, 6, 2, 4, 7, 9])
exam_scores   = np.array([55, 62, 45, 78, 70, 85, 58, 90, 95, 75, 82, 50, 68, 88, 98])
hours_slept   = np.array([9, 8, 10, 6, 7, 5, 8, 5, 4, 7, 6, 9, 7, 5, 3])

n = len(hours_studied)
```

```python
mean_studied = sum(hours_studied) / n
mean_scores = sum(exam_scores) / n
print(mean_studied, mean_scores)

mean_studied_np = np.mean(hours_studied)
mean_scores_np = np.mean(exam_scores)
print(mean_studied_np, mean_scores_np)
```

```python
var_studied = sum((i - mean_studied) ** 2 for i in hours_studied) / (n - 1)
var_scores = sum((i - mean_scores) ** 2 for i in exam_scores) / (n - 1)
print(var_studied, var_scores)

var_studied_np = np.var(hours_studied, ddof=1)
var_scores_np = np.var(exam_scores, ddof=1)
print(var_studied_np, var_scores_np)
```

```python
cov_studied_scores = sum((hours_studied[i] - mean_studied) * (exam_scores[i] - mean_scores) for i in range(n)) / (n - 1)
print(cov_studied_scores)

cov_studied_scores_np = np.cov(hours_studied, exam_scores)[0][1] # [[variance(x), cov(x, y)], [cov(y, x), variance(y)]]
print(cov_studied_scores_np)
```

```python
corr_studied_scores = cov_studied_scores / (np.sqrt(var_studied) * np.sqrt(var_scores))
print(corr_studied_scores)

corr_studied_scores_np = np.corrcoef(hours_studied, exam_scores)[0][1]
print(corr_studied_scores_np)
```
