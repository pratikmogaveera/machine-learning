```python
import random
import math
import numpy as np
```

```python
n = 100
arr = [random.randint(1, 1000) for i in range(n)]
print(arr)
```

```python
mean = sum(arr) / n
print(mean, np.mean(arr))
```

```python
sortedArr = sorted(arr)
median = (sortedArr[n//2 - 1] + sortedArr[n//2]) / 2 if n%2==0 else sortedArr[n//2]
print(median, np.median(arr))
```

```python
variance = sum((i - mean) ** 2 for i in arr) / (n - 1)
print(variance, np.var(arr, ddof=1))

std = math.sqrt(variance)
print(std, np.std(arr, ddof=1))
```
